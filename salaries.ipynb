```python
import pandas as pd
import matplotlib.pyplot as plt

salaries = pd.read_csv('Latest_Data_Science_Salaries.csv')

display(salaries.describe())

display(salaries.info())

display(salaries.head(10))

display(salaries.shape)
```

```python

# Define the mapping dictionary
level_mapping = {
    'Entry': 1,
    'Mid': 2,
    'Senior': 3,
    'Executive': 4
}

# Create the new column "Experience Level as int" using the mapping
salaries['Experience Level as int'] = salaries['Experience Level'].map(level_mapping)

# Print the updated DataFrame
print(salaries)
```

```python
columns_to_analyze = []

# Compute the correlation matrix
corr_matrix = salaries[columns_to_analyze].corr()

# Iterate over the correlation matrix and print the correlated categories
for col1 in corr_matrix.columns:
    for col2 in corr_matrix.columns:
        if col1 != col2:
            correlation = corr_matrix.loc[col1, col2]
            if correlation >= 0.5 or correlation <= -0.5:
                print(f"Correlation between {col1} and {col2}: {correlation}")
```
